### It does appear that we have sufficient data from the years of 1999 to 2025 which is the most recent completed season

```python
# View count of total 'Tournament Winner?_ref' and 'Season'
tournament_winners = df[df['tournament_winner'] == 'Yes']
tournament_winners_count = tournament_winners['season'].value_counts().sort_index()
print(tournament_winners_count)
```

### There are more than 1 tournament winner for the years 2002, 2003, 2013 and also no winner in 2025

```python
# View all columns where 'tournament winner' is 'Yes' and 'Season' is 2002, 2003, 2013
tournament_winners_2002 = df[(df['tournament_winner'] == 'Yes') & (df['season'] == 2002)]
tournament_winners_2003 = df[(df['tournament_winner'] == 'Yes') & (df['season'] == 2003)]
tournament_winners_2013 = df[(df['tournament_winner'] == 'Yes') & (df['season'] == 2013)]
```

```python
tournament_winners_2002
```

```python
# After inspection, the 2002 winner Maryland had two rows for the same season, the only difference being on had Conference as 'ACC' and the other had 'Big Ten'. This is likely a data entry error, so we will drop the row with 'Big Ten' since Maryland was in the ACC in 2002.
# Drop the row with 'Big Ten' for Maryland in 2002
df = df[~((df['season'] == 2002) & (df['tournament_winner'] == 'Yes') & (df['Conference'] == 'Big Ten'))]
```

```python
tournament_winners_2003
```

```python
# Once again in 2003, the winner Syracuse had two rows for the same season, the only difference being on had Conference as 'BE' and the other had 'ACC'. This is likely a data entry error, so we will drop the row with 'ACC' since Syracuse was in the Big East in 2003.
df = df[~((df['season'] == 2003) & (df['tournament_winner'] == 'Yes') & (df['Conference'] == 'ACC'))]
```

```python
tournament_winners_2013
```

```python
# In 2013, the winner Louisville had four rows for the same season, the only difference being on had Conference as 'ACC', 'Amer', 'BE', and 'CUSA' respectively. This is likely a data entry error, so we will drop the rows with 'ACC', 'Amer', and 'CUSA' since Louisville was in the BE in 2013.
df = df[~((df['season'] == 2013) & (df['tournament_winner'] == 'Yes') & (df['Conference'].isin(['ACC', 'Amer', 'CUSA'])))]
```

```python
# Check where season = 2025 since there was no winner listed for that season
df[df['season'] == 2025]
```

```python
# Show duplicates for combining season and team_name and Conference_extra
duplicates = df[df.duplicated(subset=['season', 'team_name', 'Conference_extra'], keep=False)]
print(f"Number of duplicate rows: {duplicates.shape[0]}")
```

```python
# Print start of duplicates
duplicates.head(25)
```

## After further inspection, there are two columns that represent Conference; 'Conference' & 'Conference_extra'
## For the values that had multiple columns under their team, the 'Conference' value was the one with multiple various values and 'Conference_extra' had the correct value for each team. It has been seemed necessary to remove 'Conference' and rename 'Conference_extra' to 'conference'

```python
# Drop 'Conference' and keep 'Conference_extra'
df.drop(columns=['Conference'], inplace=True)
# Rename 'Conference_extra' to 'Conference'
df.rename(columns={'Conference_extra': 'Conference'}, inplace=True)
```

```python
# Find full duplicates
full_duplicates = df[df.duplicated(keep=False)]
print(f"Number of full duplicate rows: {full_duplicates.shape[0]}")
```

```python
full_duplicates.head()
```

```python
# Remove full duplicates
df.drop_duplicates(inplace=True)
```

```python
# Check duplicates again after dropping
full_duplicates_after = df[df.duplicated(keep=False)]
print(f"Number of full duplicate rows after dropping: {full_duplicates_after.shape[0]}")
```

```python
# View count of total 'Tournament Winner?_ref' and 'Season'
tournament_winners = df[df['tournament_winner'] == 'Yes']
tournament_winners_count = tournament_winners['season'].value_counts().sort_index()
print(tournament_winners_count)
```

## Change Columns for next steps

```python
# Remove _ref from 'Post-Season Tournament_ref' & 'Tournament Championship?_ref' & 'Final Four?_ref' & 'Top 12 in AP Top 25 During Week 6?_ref'
df.rename(columns={'Post-Season Tournament_ref': 'Post-Season Tournament', 
                   'Tournament Championship?_ref': 'Tournament Championship?', 
                   'Final Four?_ref': 'Final Four?',
                   'Top 12 in AP Top 25 During Week 6?_ref': 'Top 12 in AP Top 25 During Week 6?'}, inplace=True)
```

```python
# Check all columns
df.columns
```

```python
# Remove _extra from all columns that have it
df.rename(columns=lambda x: x.replace('_extra', '') if '_extra' in x else x, inplace=True)
```

```python
# Rename all columns to undercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.columns
```

```python
# Final shape of dataframe after cleaning
df.shape
```

## Impute values for null in boolean columns

```python
# Impute missing values in 'tournament_winner', 'tournament_championship?', 'final_four?', 'top_12_in_ap_top_25_during_week_6?' with 'No'
cols_to_impute = ['tournament_winner', 'tournament_championship?', 'final_four?', 'top_12_in_ap_top_25_during_week_6?']
for col in cols_to_impute:
    df[col] = df[col].fillna('No')
```
