# Addressing missing values and unneccessary data


## Remove duplicate columns

```python
# Identify duplicate base column names
cols = full_kenpom_df.columns
base_names = {}

for col in cols:
    if col.endswith('_x') or col.endswith('_y'):
        base = col[:-2]
        base_names.setdefault(base, []).append(col)

# Compare missing counts and drop worse version
for base, variants in base_names.items():
    if len(variants) == 2:
        col1, col2 = variants
        if full_kenpom_df[col1].isna().sum() <= full_kenpom_df[col2].isna().sum():
            full_kenpom_df.drop(columns=col2, inplace=True)
            full_kenpom_df.rename(columns={col1: base}, inplace=True)
        else:
            full_kenpom_df.drop(columns=col1, inplace=True)
            full_kenpom_df.rename(columns={col2: base}, inplace=True)
```

## Drop rank columns and not needed columns

```python
# Drop rank columns
rank_cols = [col for col in full_kenpom_df.columns if 'Rank' in col]
full_kenpom_df.drop(columns=rank_cols, inplace=True)
```

```python
# Unnec columns
Unnec_columns = leakage_cols = [
    'Vulnerable Top 2 Seed?_ref',
    'Seed_ref',
    'Region_ref',
    'Post-Season Tournament Sorting Index_ref',
    'Mapped Conference Name',
    'Mapped ESPN Team Name'
]

full_kenpom_df.drop(columns=[col for col in leakage_cols if col in full_kenpom_df.columns], inplace=True)
```

## Impute median for remaining statistical values

```python
numeric_cols = full_kenpom_df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    if full_kenpom_df[col].isna().sum() > 0:
        full_kenpom_df[col].fillna(full_kenpom_df[col].median(), inplace=True)
```

```python
# Check changes

# Explore missing data in full_kenpom_df
missing_data = full_kenpom_df.isnull().sum()
missing_data = missing_data[missing_data > 0].sort_values(ascending=False)
print("Columns with missing data in KenPom and Reference Data:")
print(missing_data)
```
