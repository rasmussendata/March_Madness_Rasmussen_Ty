# 2. Rename and merge relevant columns

```python
# KP def minimum year
min_year = kp_def['Season'].min()
print(f'Minimum Year in KenPom Defense Data: {min_year}')
```

```python
kp_def.head()
```

```python
# Rename offensive columns to _off
kp_off = kp_off.add_suffix('_off')
```

```python
# Rename defensive columns to _def
kp_def = kp_def.add_suffix('_def')
```

```python
# Change season_def and TeamName_def to season and TeamName for merging
kp_def = kp_def.rename(columns={'Season_def': 'Season', 'TeamName_def': 'TeamName'})
kp_off = kp_off.rename(columns={'Season_off': 'Season', 'TeamName_off': 'TeamName'})
```

```python
kp_def.head()
```

## Merge all data from 2002-2025 dataset into one

```python
# Merge dataframes
full_df = pd.merge(
kp_off,
kp_def,
on=['Season', 'TeamName'],
how='outer'
)
```

```python
full_df.head()
```

```python
kp_eff.head()
```

```python
# Rename kp_eff Team to TeamName for merging
kp_eff = kp_eff.rename(columns={'Team': 'TeamName'})
```

```python
kp_mis.head()
```

```python
kp_pointdis.head()  
```

```python
kp_sumpre.head()
```

```python
kp_sum.head()
```

```python
# Merge all KenPom dataframes into one
full_kenpom_df = full_df.copy()
dataframes = [kp_eff, kp_hei, kp_mis, kp_pointdis, kp_sumpre, kp_sum]
for df in dataframes:
    # Rename columns to avoid conflicts
    df_renamed = df.add_suffix('_extra')
    df_renamed = df_renamed.rename(columns={'Season_extra': 'Season', 'TeamName_extra': 'TeamName'})
    
    # Merge with the full_kenpom_df
    full_kenpom_df = pd.merge(
        full_kenpom_df,
        df_renamed,
        on=['Season', 'TeamName'],
        how='outer'
    )
full_kenpom_df.head()
```

### # Merge REF Post season and Top 12 dataframes


```python
ref_tour_teams.head()
```

```python
# Rename Team Name to TeamName for merging
ref_tour_teams = ref_tour_teams.rename(columns={'Team Name': 'TeamName'})
```

```python
ref_top12_week6.head()
```

```python
# Rename Mapped ESPN Team Name to TeamName for merging
ref_top12_week6 = ref_top12_week6.rename(columns={'Mapped ESPN Team Name': 'TeamName'})
```

```python
# Add two ref tables to full_kenpom_df
ref_dataframes = [ref_tour_teams, ref_top12_week6]
for ref_df in ref_dataframes:
    # Rename columns to avoid conflicts
    ref_df_renamed = ref_df.add_suffix('_ref')
    ref_df_renamed = ref_df_renamed.rename(columns={'Season_ref': 'Season', 'TeamName_ref': 'TeamName'})
    
    # Merge with the full_kenpom_df
    full_kenpom_df = pd.merge(
        full_kenpom_df,
        ref_df_renamed,
        on=['Season', 'TeamName'],
        how='outer'
    )
```

```python
# Check ref_conf table for column name
ref_conf.head()
```

```python
# Add conference columns
full_kenpom_df = pd.merge(
    full_kenpom_df,
    ref_conf,
    on='TeamName',
    how='left'
)
```

```python
# Check new dataframe that includes all data for 2002-2025 seasons
full_kenpom_df.head()
```
